```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#1.Top 7 RICE PRODUCTION State Data(Bar_plot)
#
total_area = df['RICE AREA (1000 ha)'].sum()
total_production = df['RICE PRODUCTION (1000 tons)'].sum()
average_yield = df['RICE YIELD (Kg per ha)'].mean()
print(f"Total Rice Area: {total_area} (1000 ha)")
print(f"Total Rice Production: {total_production} (1000 tons)")
print(f"Average Rice Yield: {average_yield:.2f} Kg/ha")

top_states = df.groupby('State Code')['RICE PRODUCTION (1000 tons)'].sum().sort_values(ascending=False).head(7)

top_states.plot(kind='bar', color='green')
plt.title("Top 7 Rice Producing States in India")
plt.ylabel("Production (1000 tons)")
plt.xlabel("State")
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()



```

```python
#2.Top 5 Wheat Producing States Data(Bar_chart)and its percentage(%)(Pie_chart)
#
wheat_data = df.groupby('State Code')[['WHEAT PRODUCTION (1000 tons)']].sum()
top5_wheat = wheat_data.sort_values(by='WHEAT PRODUCTION (1000 tons)', ascending=False).head(5)

plt.figure(figsize=(8, 5))
top5_wheat.plot(kind='bar', legend=False)
plt.title("Top 5 Wheat Producing States")
plt.ylabel("Production (1000 tons)")
plt.xlabel("State")
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()

plt.figure(figsize=(6, 6))
plt.pie(top5_wheat['WHEAT PRODUCTION (1000 tons)'], labels=top5_wheat.index, autopct='%1.1f%%', startangle=140)
plt.title("Percentage Share of Top 5 Wheat Producing States")
plt.show()
```

```python
#3.Oil seed production by top 5 states#
oilseed_data = df.groupby('State Code')[['OILSEEDS PRODUCTION (1000 tons)']].sum()
top5_oilseed = oilseed_data.sort_values(by='OILSEEDS PRODUCTION (1000 tons)', ascending=False).head(5)

plt.figure(figsize=(8,5))
top5_oilseed.plot(kind='bar', legend=False)
plt.title('Top 5 Oil Seed Producing States')
plt.ylabel('Production (1000 tons)')
plt.xlabel('State Code')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()

```

```python
#4.Top 7 SUNFLOWER PRODUCTION  State#
sunflower_data = df.groupby('State Code')[['SUNFLOWER PRODUCTION (1000 tons)']].sum()
top7_sunflower = sunflower_data.sort_values(by='SUNFLOWER PRODUCTION (1000 tons)', ascending=False).head(7)

plt.figure(figsize=(10,6))
top7_sunflower.plot(kind='bar', legend=False, color='orange')
plt.title('Top 7 Sunflower Producing States')
plt.ylabel('Production (1000 tons)')
plt.xlabel('State Code')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()


```

```python
#5.India's SUGARCANE PRODUCTION From Last 50 Years(Line_plot)#
last_50_years = df['Year'] >= (df['Year'].max() - 49)  

sugarcane_data = df[last_50_years].groupby('Year')[['SUGARCANE PRODUCTION (1000 tons)']].sum()
plt.figure(figsize=(12,6))
plt.plot(sugarcane_data.index, sugarcane_data['SUGARCANE PRODUCTION (1000 tons)'], marker='o', color='green')
plt.title("India's Sugarcane Production - Last 50 Years")
plt.xlabel('Year')
plt.ylabel('Production (1000 tons)')
plt.grid(True)
plt.show()


```

```python
#6.Rice Production Vs Wheat Production (Last 50y)#

last_50_years = df['Year'] >= (df['Year'].max() - 49)

rice_prod = df[last_50_years].groupby('Year')['RICE PRODUCTION (1000 tons)'].sum()
wheat_prod = df[last_50_years].groupby('Year')['WHEAT PRODUCTION (1000 tons)'].sum()

plt.figure(figsize=(12,6))
plt.plot(rice_prod.index, rice_prod.values, label='Rice Production', color='blue', marker='o')
plt.plot(wheat_prod.index, wheat_prod.values, label='Wheat Production', color='orange', marker='s')
plt.title("Rice vs Wheat Production in Last 50 Years")
plt.xlabel("Year")
plt.ylabel("Production (1000 tons)")
plt.legend()
plt.grid(True)
plt.show()



```

```python
#7.Rice Production By West Bengal Districts#

wb_rice = df[df['State Name'].str.lower() == 'west bengal']
wb_rice_prod = wb_rice.groupby('Dist Code')['RICE PRODUCTION (1000 tons)'].sum().sort_values(ascending=False)

plt.figure(figsize=(12,6))
wb_rice_prod.plot(kind='bar', color='green')
plt.title("Rice Production by Districts in West Bengal")
plt.xlabel("District")
plt.ylabel("Production (1000 tons)")
plt.xticks(rotation=0, ha='right')
plt.tight_layout()
plt.show()
```

```python
#8.Top 10 Wheat Production Years From UP#

up_wheat = df[df['State Name'].str.lower() == 'uttar pradesh']
up_wheat_prod = up_wheat.groupby('Year')['WHEAT PRODUCTION (1000 tons)'].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(10,6))
up_wheat_prod.sort_index().plot(kind='bar', color='orange')
plt.title("Top 10 Wheat Production Years - Uttar Pradesh")
plt.xlabel("Year")
plt.ylabel("Wheat Production (1000 tons)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#9.Millet Production (Last 50y)#

millet_data = df.groupby('Year')[['PEARL MILLET PRODUCTION (1000 tons)', 'FINGER MILLET PRODUCTION (1000 tons)']].sum().reset_index()

plt.figure(figsize=(12, 6))
plt.plot(millet_data['Year'], millet_data['PEARL MILLET PRODUCTION (1000 tons)'], label='Pearl Millet', marker='o')
plt.plot(millet_data['Year'], millet_data['FINGER MILLET PRODUCTION (1000 tons)'], label='Finger Millet', marker='s')

plt.title('Pearl & Finger Millet Production in India (Last 50 Years)')
plt.xlabel('Year')
plt.ylabel('Production (1000 tons)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

```

```python
#10.Sorghum Production (Kharif and Rabi) by Region#

sorghum_data = df.groupby('State Code')[['KHARIF SORGHUM PRODUCTION (1000 tons)','RABI SORGHUM PRODUCTION (1000 tons)']].sum().reset_index()
sorghum_data['Total'] = sorghum_data['KHARIF SORGHUM PRODUCTION (1000 tons)'] + sorghum_data['RABI SORGHUM PRODUCTION (1000 tons)']
top_regions = sorghum_data.sort_values(by='Total', ascending=False).head(10)

plt.figure(figsize=(14, 6))
bar_width = 0.4
r1 = range(len(top_regions))
r2 = [x + bar_width for x in r1]

plt.bar(r1, top_regions['KHARIF SORGHUM PRODUCTION (1000 tons)'], width=bar_width, label='Kharif')
plt.bar(r2, top_regions['RABI SORGHUM PRODUCTION (1000 tons)'], width=bar_width, label='Rabi')

plt.xlabel('Region')
plt.ylabel('Production (1000 tons)')
plt.title('Top 10 Sorghum Producing Regions - Kharif vs Rabi')
plt.xticks([r + bar_width/2 for r in range(len(top_regions))], top_regions['State Code'], rotation=0)
plt.legend()
plt.tight_layout()
plt.grid(True)
plt.show()
```

```python
#11.Top 7 States for Groundnut Production#

groundnut_data = df.groupby('State Code')['GROUNDNUT PRODUCTION (1000 tons)'].sum().reset_index()
top_groundnut = groundnut_data.sort_values(by='GROUNDNUT PRODUCTION (1000 tons)', ascending=False).head(7)

plt.figure(figsize=(10,6))
plt.bar(top_groundnut['State Code'], top_groundnut['GROUNDNUT PRODUCTION (1000 tons)'], color='sandybrown')
plt.xlabel('State')
plt.ylabel('Production (1000 tons)')
plt.title('Top 7 States for Groundnut Production')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()

```

```python
#12.Soybean Production by Top 5 States and Yield Efficiency#

soyabean_data = df.groupby('State Code').agg({
    'SOYABEAN PRODUCTION (1000 tons)': 'sum',
    'SOYABEAN YIELD (Kg per ha)': 'mean'
}).reset_index()

top_soyabean = soyabean_data.sort_values(by='SOYABEAN PRODUCTION (1000 tons)', ascending=False).head(5)

fig, ax1 = plt.subplots(figsize=(10,6))
ax1.bar(top_soyabean['State Code'], top_soyabean['SOYABEAN PRODUCTION (1000 tons)'], color='olive', label='Production')
ax1.set_xlabel('State')
ax1.set_ylabel('Production (1000 tons)', color='olive')
ax1.tick_params(axis='y', labelcolor='olive')

ax2 = ax1.twinx()
ax2.plot(top_soyabean['State Code'], top_soyabean['SOYABEAN YIELD (Kg per ha)'], color='darkred', marker='o', label='Yield')
ax2.set_ylabel('Yield (Kg per ha)', color='darkred')
ax2.tick_params(axis='y', labelcolor='darkred')

plt.title('Top 5 States for Soyabean Production and Yield Efficiency')
fig.tight_layout()
plt.show()
```

```python
#13.Oilseed Production in Major States#

oilseed_prod = df.groupby('State Code')['OILSEEDS PRODUCTION (1000 tons)'].sum().sort_values(ascending=False).head(7)

plt.figure(figsize=(10,6))
oilseed_prod.plot(kind='bar', color='seagreen')
plt.title('Top 7 States in Oilseed Production')
plt.xlabel('State')
plt.ylabel('Production (1000 tons)')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
#14.Impact of Area Cultivated on Production (Rice, Wheat, Maize)#

plt.figure(figsize=(12,6))
plt.scatter(df['RICE AREA (1000 ha)'], df['RICE PRODUCTION (1000 tons)'], color='green', label='Rice', alpha=0.6)
plt.scatter(df['WHEAT AREA (1000 ha)'], df['WHEAT PRODUCTION (1000 tons)'], color='orange', label='Wheat', alpha=0.6)
plt.scatter(df['MAIZE AREA (1000 ha)'], df['MAIZE PRODUCTION (1000 tons)'], color='blue', label='Maize', alpha=0.6)

plt.xlabel('Area Cultivated (1000 ha)')
plt.ylabel('Production (1000 tons)')
plt.title('Impact of Area Cultivated on Production (Rice, Wheat, Maize)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

```

```python
#15.Rice vs. Wheat Yield Across States#

yield_df = df.groupby('State Code')[['RICE YIELD (Kg per ha)', 'WHEAT YIELD (Kg per ha)']].mean().dropna()

yield_df.plot(kind='bar', figsize=(12,6))
plt.title("Rice vs. Wheat Yield Across States")
plt.ylabel("Yield (Kg per ha)")
plt.xlabel("State")
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()


```
